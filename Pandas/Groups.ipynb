```python
evodf = pd.merge(df22jan,df23jan,on='Carrier')
```

```python
evodf.columns = ['Carrier','22 jan', '23 jan']
```

```python
evodf
```

```python
evodf['Evolution'] = 100 * (evodf['23 jan']-evodf['22 jan'])/evodf['22 jan']
```

```python
evodf['percent'] = np.round(100 * (evodf['23 jan']-evodf['22 jan'])/evodf['22 jan'],2)
evodf
```

```python
def dec2(n):
    return round(n,2)
```

```python
evodf["Evolution2"] = evodf["Evolution"].apply(dec2)
```
